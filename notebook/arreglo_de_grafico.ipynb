```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
# Corrección: usar StringIO de 'io' en lugar de 'pd.compat.StringIO' para leer el CSV.
df = pd.read_csv(StringIO(data))

# Convertir la columna 'fecha' a formato datetime
# Corrección: el nombre de la columna es 'fecha', no 'fechas'.
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
# Gráfico de líneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o')  # Añadimos marcadores para claridad.
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid()  # Añadimos una cuadrícula para mejorar la legibilidad.
plt.show()
```

```python
# Crear un gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='red')
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid()  # Añadimos una cuadrícula.
plt.show()
```

```python
# Crear un histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribución de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid()  # Añadimos una cuadrícula.
plt.show()
```
